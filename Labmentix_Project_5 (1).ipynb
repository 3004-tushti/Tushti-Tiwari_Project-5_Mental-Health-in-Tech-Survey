## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# 1. Copy the original DataFrame
df_clean = df.copy()

# 2. Remove unrealistic ages
df_clean = df_clean[(df_clean['Age'] >= 15) & (df_clean['Age'] <= 100)]

# 3. Standardize gender values
df_clean['Gender'] = df_clean['Gender'].astype(str).str.lower().str.strip()

# Define gender keyword groups
male_keywords = ['male', 'm', 'man', 'cis male', 'msle', 'mal']
female_keywords = ['female', 'f', 'woman', 'cis female', 'femail']
other_keywords = ['trans', 'non-binary', 'queer', 'genderfluid', 'genderqueer', 'agender', 'other']

# Gender normalization function
def clean_gender(gender):
    for word in male_keywords:
        if word in gender:
            return 'Male'
    for word in female_keywords:
        if word in gender:
            return 'Female'
    for word in other_keywords:
        if word in gender:
            return 'Other'
    return 'Other'

# Apply gender cleaning
df_clean['Gender'] = df_clean['Gender'].apply(clean_gender)

# 4. Handle missing values
fill_with_unknown = ['self_employed', 'work_interfere']
for col in fill_with_unknown:
    df_clean[col] = df_clean[col].fillna('Unknown')

# Drop unnecessary columns (sparse or free text)
df_clean.drop(columns=['comments', 'state'], inplace=True)

# 5. Drop duplicates
df_clean = df_clean.drop_duplicates().reset_index(drop=True)

# 6. Final confirmation
print("\n✅ Cleaned dataset is ready for analysis.\n")
df_clean.info()
````

### What all manipulations have you done and insights you found?

**What all manipulations have you done and insights you found?**

While preparing the dataset for analysis, I performed a series of cleaning and preprocessing steps to ensure the data was accurate, consistent, and ready for meaningful insights.

1. Age Filtering:
I removed entries with unrealistic age values (e.g., under 15 or over 100). These outliers could skew the results, so only respondents within a valid working age range were retained.

2. Gender Standardization:
The Gender column had a wide variety of inconsistent inputs like “M”, “male-ish”, “cis female”, etc. I grouped these into three clean categories: Male, Female, and Other. This helped create reliable gender-based comparisons later.

3. Handling Missing Values:
Some key columns like self_employed and work_interfere had missing data. Instead of dropping these records, I filled the missing values with the label 'Unknown' to retain them for analysis without losing information.

4. Dropping Irrelevant Columns:
I dropped the comments and state columns. The comments field was mostly empty and not useful for structured analysis, and state was relevant only to U.S. respondents and had many missing values.

5. Duplicate Removal:
I checked for and removed any duplicate records to ensure each response was unique and to maintain data integrity.

6. Index Reset:
After filtering, I reset the DataFrame index to keep the structure clean and continuous.

**Key Early Insights**

1. The dataset originally had some messy and inconsistent data, especially in the Gender and Age columns.

2. Many respondents skipped optional questions like work_interfere, which may indicate discomfort or lack of clarity around mental health in the workplace.

3. The cleaning process helped reduce noise and focus on real, usable responses — bringing the dataset down to around 1,250 clean and valid entries.

####  Objective 2- Chart 1: Top 10 Countries by Mental Health Treatment Rate

````python
# Objective 1- Chart 1: Top 10 Countries by Mental Health Treatment Rate
# Filter top 10 countries by response count
top_countries = df_clean['Country'].value_counts().head(10).index
df_top = df_clean[df_clean['Country'].isin(top_countries)]

# Calculate treatment rate per country (in %)
treatment_rate = df_top.groupby('Country')['treatment'].value_counts(normalize=True).unstack().fillna(0) * 100

# Plot bar chart for 'Yes' responses
plt.figure(figsize=(9,6))
treatment_rate['Yes'].sort_values().plot(kind='barh', color='mediumseagreen')
plt.title("Top 10 Countries by Mental Health Treatment Rate", fontsize=14)
plt.xlabel("Percentage of Respondents Who Sought Treatment")
plt.ylabel("Country")
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.show()
````

#### Objective 2- Chart 2: Global Choropleth Map – Mental Health Treatment Rate by Country

````python
# Objective 2- Chart 2: Global Choropleth Map – Mental Health Treatment Rate by Country
import plotly.express as px

# Prepare data: treatment rate by country
country_treatment = df_clean.groupby('Country')['treatment'].value_counts(normalize=True).unstack().fillna(0) * 100
country_treatment = country_treatment.reset_index()

# Rename column for clarity
country_treatment.rename(columns={'Yes': 'Treatment Rate (%)'}, inplace=True)

# Plot interactive map
fig = px.choropleth(
    country_treatment,
    locations='Country',
    locationmode='country names',
    color='Treatment Rate (%)',
    hover_name='Country',
    color_continuous_scale='Tealgrn',
    title='Global Mental Health Treatment Rate by Country (%)'
)

fig.show()
````

#### Objective 4- Chart 4: Radar Chart – Mental Health Interference by Age Group

````python
# Objective 4- Chart 4: Radar Chart – Mental Health Interference by Age Group
# Step 1: Create age group bins
df_radar = df_clean.copy()
df_radar['age_group'] = pd.cut(df_radar['Age'], bins=[15, 24, 34, 44, 54, 100],
                               labels=['15–24', '25–34', '35–44', '45–54', '55+'])

# Step 2: Calculate % of each 'work_interfere' level per age group
interfere_dist = df_radar.groupby('age_group')['work_interfere'].value_counts(normalize=True).unstack().fillna(0) * 100

# Step 3: Plot radar chart (matplotlib)
import numpy as np

labels = interfere_dist.columns.tolist()
num_vars = len(labels)

angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]  # close the circle

# Radar plot
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

for age_group in interfere_dist.index:
    values = interfere_dist.loc[age_group].tolist()
    values += values[:1]  # complete the loop
    ax.plot(angles, values, label=age_group)
    ax.fill(angles, values, alpha=0.1)

ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)
ax.set_thetagrids(np.degrees(angles[:-1]), labels)
ax.set_title("Work Interference by Age Group", fontsize=14)
ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))
plt.show()
````

##### 1. Why did you pick the specific chart?

A radar chart lets us compare multiple response categories (work_interfere levels) across multiple groups (age_group) on a single, circular scale — making it easy to spot which age group is most affected by workplace mental health interference.

##### 2. What is/are the insight(s) found from the chart?

#### Objective 7- Chart 3: Radial Bar Chart – Treatment Rate by Family History

````python
# Objective 7- Chart 3: Radial Bar Chart – Treatment Rate by Family History
# Prepare the data
treatment_counts = df_clean.groupby('family_history')['treatment'].value_counts(normalize=True).unstack().fillna(0) * 100
treatment_counts = treatment_counts.reset_index()
treatment_counts.columns.name = None

# Flatten for radial chart
labels = []
values = []
colors = []

color_map = {'Yes': 'mediumseagreen', 'No': 'indianred'}

for i, row in treatment_counts.iterrows():
    for col in ['Yes', 'No']:
        labels.append(f"{row['family_history']} - {col}")
        values.append(row[col])
        colors.append(color_map[col])

# Calculate angles
N = len(values)
angles = [n / float(N) * 2 * np.pi for n in range(N)]

# Plot
fig = plt.figure(figsize=(10, 10))
ax = plt.subplot(111, polar=True)
bars = ax.bar(angles, values, width=0.4, color=colors, edgecolor='black')

# Add labels
for i, (bar, label) in enumerate(zip(bars, labels)):
    angle = angles[i]
    rotation = np.degrees(angle)
    alignment = 'left' if angle < np.pi else 'right'
    ax.text(angle, bar.get_height() + 2, label, ha=alignment, va='center', rotation=rotation if angle < np.pi else rotation + 180)

ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)
ax.set_ylim(0, max(values) + 10)
ax.set_yticklabels([])
ax.set_xticklabels([])
plt.title("Treatment Rate by Family History", fontsize=15)
plt.show()
````

##### 1. Why did you pick the specific chart?

A radial bar chart adds visual storytelling power while keeping data proportion clarity. It's ideal for highlighting contrasting treatment-seeking behavior based on family history — and adds aesthetic value to final deliverables like slides and dashboards.

##### 2. What is/are the insight(s) found from the chart?

1. Those with a family history of mental illness are far more likely to seek treatment.

2. Respondents without a family history have significantly lower treatment rates.

3. Visually, the difference is large and immediate — reinforcing that family background is a strong predictor of help-seeking behavior.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

**POSITIVE INSIGHT:**

1. Supports early screening or optional disclosure systems to help identify high-risk employees and offer discreet assistance.

2. Also informs targeted awareness programs to encourage treatment among those without family history (who might otherwise downplay symptoms).

**NEGATIVE INSIGHT:**

1. Over-reliance on family history as a marker may exclude individuals struggling without such a background — so companies must ensure inclusive, accessible support for all.
